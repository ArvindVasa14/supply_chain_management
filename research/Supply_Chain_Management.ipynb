### Outlier check

```python
# comparing mean and median of potential outliers
data.describe().loc[['mean','50%'],:]
```

```python
# count of outliers and percentage of outliers
print("Feature"+" "*23+"outlier count"+" "*5+"outlier percentage" )
print("-"*70)
outlier_containing_features=[]
for feature in data.select_dtypes(['int','float']):
    outlier_count= 0
    Q1= data[feature].quantile(0.25)
    Q3= data[feature].quantile(0.75)
    IQR= Q3 - Q1
    outlier_count= (data[feature]>(Q3+1.5*IQR)).sum() + (data[feature]<(Q1-1.5*IQR)).sum()
    if(outlier_count>0):
        outlier_containing_features.append(feature)
        print(feature," "*(32-len(feature)), outlier_count," "*(20-len(str(outlier_count))), 
          f'{((outlier_count/25000)*100):.2f} %')
```

```python
# plotting box plot to visualize the outliers
plt.subplots(figsize=(12,12))
for i,feature in enumerate(outlier_containing_features):
    plt.subplot(3,3,i+1)
    plt.xlabel(feature)
    sns.boxplot(data[feature])
```

### Outlier Treatment

```python
# Feature                       outlier count     outlier percentage
# ----------------------------------------------------------------------
# transport_issue_l1y               2943                  11.77 %
# Competitor_in_mkt                 96                    0.38 %
# retail_shop_num                   948                   3.79 %
# workers_num                       607                   2.43 %
```

```python
# This method aims to cap extreme values by setting them to the nearest reasonable boundary within the 1.5 * IQR range

data_treated= data.copy()
for feature in ['transport_issue_l1y','Competitor_in_mkt','retail_shop_num','workers_num']:
    Q1= data[feature].quantile(0.25)
    Q3= data[feature].quantile(0.75)
    IQR= Q3 - Q1
    lower_limit= Q1-1.5*IQR
    upper_limit= Q3+1.5*IQR
    data_treated.loc[data_treated[feature]<lower_limit,feature]= lower_limit
    data_treated.loc[data_treated[feature]>upper_limit,feature]= upper_limit
```

```python
print("Feature"+" "*23+"outlier count"+" "*5+"outlier percentage" )
print("-"*70)
outlier_containing_features=[]
for feature in ['transport_issue_l1y','Competitor_in_mkt','retail_shop_num','workers_num']:
    outlier_count= 0
    Q1= data_treated[feature].quantile(0.25)
    Q3= data_treated[feature].quantile(0.75)
    IQR= Q3 - Q1
    outlier_count= (data_treated[feature]>(Q3+1.5*IQR)).sum() + (data_treated[feature]<(Q1-1.5*IQR)).sum()
    if(outlier_count>0):
        outlier_containing_features.append(feature)
        print(feature," "*(32-len(feature)), outlier_count," "*(20-len(str(outlier_count))), 
          f'{((outlier_count/25000)*100):.2f} %')
```

### PREPROCESSING

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

### Encoding the data

```python
data.select_dtypes('object').columns
```

```python
data.columns
```

#### One Hot Encoding

```python
# since the data is nominal, going with one hot enconding
data_encoded= data.copy()
nominal_features= ['Location_type','zone','WH_regional_zone','wh_owner_type']
for cate_feature in nominal_features:
    print(cate_feature)
    one_hot_encoded = pd.get_dummies(data_encoded[cate_feature], prefix=cate_feature)
    data_encoded.drop(cate_feature, axis=1, inplace=True)
    data_encoded= pd.concat([data_encoded, one_hot_encoded], axis=1)
    
data_encoded.head()
```

#### Label Encoding

```python
# since the data is ordinal, going with Label encoding
Le= LabelEncoder()
for cat_feature in ['WH_capacity_size','approved_wh_govt_certificate']:
    data_encoded[cat_feature]= Le.fit_transform(data_encoded[cat_feature])
data_encoded.head()
```

### Standardizing the data

```python
# bringing all features scale into same range  
scaler= StandardScaler()
data_encoded_std= pd.DataFrame(scaler.fit_transform(data_encoded.drop(['product_wg_ton'], axis=1)), columns=data_encoded.columns[0:-1])
data_encoded_std.head()
```

### train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X= data_encoded_std.drop('product_wg_ton', axis=1)
y= data['product_wg_ton']
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2, random_state=1)
```

## Model Selection

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.ensemble import RandomForestRegressor, BaggingRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.metrics import accuracy_score
```

```python
def evaluate_model(actual, predicted):
    mse = mean_squared_error(actual, predicted)
    mae = mean_absolute_error(actual, predicted)
    rmse= np.sqrt(mse)
    r2 = r2_score(actual, predicted)
    
    return r2, mae, mse, rmse
```

```python
models= {
    'LinearRegression':LinearRegression(),
    'Lasso':Lasso(max_iter=1000),
    'Ridge':Ridge(),
    'Elasticnet':ElasticNet(),
    "BaggingRegressor": BaggingRegressor(base_estimator=LinearRegression(),
                                         n_estimators=300)
}
```

```python
for model_name, model in models.items():
    model.fit(X_train, y_train)
#     y_train_pred= model.predict(X_train)
    y_pred_test= model.predict(X_test)
    r2, mae, mse, rmse= evaluate_model(y_test, y_pred_test)
    print(model_name)
    print('Model Training Performance')
    print(f"MSE: {mse:.2f}",)
    print(f"RMSE: {rmse:.2f}",)
    print(f"MAE: {mae:.2f}",)
    print(f"R2 score: {r2*100:.2f}",)
    print("=" *30)
    
```

### Considering Linear Regression for model building 

```python
Lin_Reg= LinearRegression()
Lin_Reg.fit(X_train, y_train)
y_pred_test= Lin_Reg.predict(X_test)
print(f"R2 of Linear regression : {(r2_score(y_test, y_pred_test))*100:.2f} %")
```

### Variable importance

```python
variable_importance= {}
coeff= Lin_Reg.coef_
for i, feature in enumerate(X.columns):
    variable_importance[feature]= coeff[i]

pd.DataFrame(variable_importance.items(), columns=['Feature', 'Importance']).sort_values(by='Importance', ascending= False)
```
